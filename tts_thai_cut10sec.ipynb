```python
from pythainlp import sent_tokenize, word_tokenize
```

```python
test = "โดยนักวิจัยจากสถาบันการศึกษา4แห่งคือจุฬาลงกรณ์มหาวิทยาลัยมหาวิทยาลัยเกษตรศาสตร์สถาบันเทคโนโลยีพระจอมเกล้าธนบุรีและสถาบันเทคโนโลยีพระจอมเกล้าลาดกระบังร่วมกันดำเนินงานวิจัย"

print('len :'+str(len(test)))
txt = word_tokenize(test, engine="deepcut")

print(txt)

print(len(txt))
```

```python
lst = []
for index, row in df1.iterrows():
    txt = word_tokenize(row[0], engine="deepcut")
    myList =''.join(txt)

    if len(myList) > 120:
      txt = txt[0:12]
      #print(txt)
      myList =''.join(txt)
      lst.append(myList)
    else:
      txt = txt[0:20]
      #print(txt)
      myList =''.join(txt)
      lst.append(myList)

#print(lst)

with open('new-text-tts.csv', 'w') as f:
    f.write('\n'.join(lst))
```
